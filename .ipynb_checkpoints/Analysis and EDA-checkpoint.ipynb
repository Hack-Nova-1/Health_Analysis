```python
import pandas as pd

# Load dataset
file_path = "Symptom-severity.csv"
df = pd.read_csv(file_path)   # âœ… You missed this line

# Basic info
df_info = df.info()
df_head = df.head()
df_shape = df.shape

df_info, df_shape, df_head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------
# Data Cleaning
# ----------------------------

# Check for duplicates
duplicates = df[df.duplicated()]

# Clean symptom names (remove extra spaces, lowercase)
df['Symptom'] = df['Symptom'].str.strip().str.lower()

# Verify uniqueness after cleaning
unique_count = df['Symptom'].nunique()

# ----------------------------
# Exploratory Data Analysis (EDA)
# ----------------------------

# Summary statistics
summary_stats = df['weight'].describe()

# Distribution plot of weights
plt.figure(figsize=(8,5))
sns.histplot(df['weight'], bins=10, kde=True, color='teal')
plt.title('Distribution of Symptom Severity (Weight)', fontsize=14)
plt.xlabel('Weight (Severity)')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Countplot for severity frequency
plt.figure(figsize=(8,5))
sns.countplot(x='weight', data=df, palette='viridis')
plt.title('Frequency of Each Severity Level', fontsize=14)
plt.xlabel('Severity Weight')
plt.ylabel('Number of Symptoms')
plt.show()

# Top 10 most severe symptoms
top_symptoms = df.sort_values('weight', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='weight', y='Symptom', data=top_symptoms, palette='rocket')
plt.title('Top 10 Most Severe Symptoms', fontsize=14)
plt.xlabel('Severity Weight')
plt.ylabel('Symptom')
plt.show()

duplicates, unique_count, summary_stats
```

```python
# ----------------------------
# Grouping Symptoms by Severity Range
# ----------------------------

# Define severity levels
def categorize_severity(weight):
    if weight <= 2:
        return 'Mild'
    elif 3 <= weight <= 5:
        return 'Moderate'
    else:
        return 'Severe'

df['Severity_Level'] = df['weight'].apply(categorize_severity)

# Count of each severity level
severity_counts = df['Severity_Level'].value_counts()

# Visualization of symptom severity categories
plt.figure(figsize=(7,5))
sns.barplot(x=severity_counts.index, y=severity_counts.values, palette='coolwarm')
plt.title('Symptom Count by Severity Category', fontsize=14)
plt.xlabel('Severity Level')
plt.ylabel('Number of Symptoms')
plt.show()

# ----------------------------
# Pie chart for visual balance
# ----------------------------
plt.figure(figsize=(6,6))
plt.pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%', colors=sns.color_palette('coolwarm', 3))
plt.title('Proportion of Severity Levels')
plt.show()

# Display summary
severity_counts
```

```python
import os
import warnings
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# ----------------------------
# Clustering Symptoms by Severity
# ----------------------------
os.environ["OMP_NUM_THREADS"] = "1"
warnings.filterwarnings("ignore", message="KMeans is known to have a memory leak")
# Prepare data for clustering
X = df[['weight']].values

# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using elbow method
inertia = []
K = range(1, 8)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot elbow method
plt.figure(figsize=(8,5))
plt.plot(K, inertia, 'o-', color='navy')
plt.title('Elbow Method for Optimal Cluster Count', fontsize=14)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True, alpha=0.3)
plt.show()

# ----------------------------
# Apply KMeans Clustering (choose 3 clusters as natural grouping)
# ----------------------------
kmeans_final = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans_final.fit_predict(X_scaled)

# ----------------------------
# Visualization of Clusters
# ----------------------------
plt.figure(figsize=(8,6))
sns.scatterplot(x=np.arange(len(df)), y='weight', hue='Cluster', data=df, palette='viridis', s=70)
plt.title('Symptom Clusters by Severity', fontsize=14)
plt.xlabel('Symptom Index')
plt.ylabel('Severity Weight')
plt.legend(title='Cluster')
plt.show()

# Cluster summary
cluster_summary = df.groupby('Cluster')['weight'].agg(['min', 'max', 'mean', 'count'])
cluster_summary
```
